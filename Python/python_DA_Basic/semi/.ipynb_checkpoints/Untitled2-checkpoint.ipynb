```python
'''
    1. 정상 운행한 날의 요소별 평균
'''
normal_dc=pd.pivot_table(indc1618_2, index=['year','month'],values=col)
normal_dc=normal_dc.reset_index()

normal_dc['yemo']=normal_dc['year'].map(str) + '-' + normal_dc['month'].map(str)

normal_dc_16=normal_dc.loc[normal_dc['year']==2016]
normal_dc_17=normal_dc.loc[normal_dc['year']==2017]
normal_dc_18=normal_dc.loc[normal_dc['year']==2018]
normal_dc_18
```

```python
'''
    2. 지연된 날의 요소별 평균
       == 지연 + 결항된 날(sum)의 요소별 평균
'''
delay_dc=pd.pivot_table(indc1618_1, index=['year','month'],values=col)
delay_dc=delay_dc.reset_index()

delay_dc_16=delay_dc.loc[delay_dc['year']==2016]
delay_dc_17=delay_dc.loc[delay_dc['year']==2017]
delay_dc_18=delay_dc.loc[delay_dc['year']==2018]
delay_dc_18
```

```python
'''
    3. 결항된 날의 요소별 평균
       == 지연 + 결항된 날(difference)의 요소별 평균
           2016: 2 3 4 5 10
           2017: 9 12
           2018: 1 3 4 5 6 8
'''
cancel_dc=pd.pivot_table(indc1618_1.loc[indc1618_1['cancel_sum']>=1],
                         index=['year','month'],values=col)
cancel_dc=cancel_dc.reset_index()

cancel_dc_16=cancel_dc.loc[cancel_dc['year']==2016]
cancel_dc_17=cancel_dc.loc[cancel_dc['year']==2017]
cancel_dc_18=cancel_dc.loc[cancel_dc['year']==2018]
cancel_dc_18
```

```python
plt.figure(figsize = (18,15))

plt.subplot(311)
plt.plot(normal_dc_16['month'],
         normal_dc_16['CO'],linewidth='3')
plt.plot(delay_dc_16['month'],
         delay_dc_16['CO'],'.g',markersize=15)
plt.plot(cancel_dc_16['month'],
         cancel_dc_16['CO'],'.m',markersize=15)
plt.title('2016 temp')
plt.legend(['normal_temp','delay_temp','cancel_temp'],prop={'size':15})

plt.subplot(312)
plt.plot(normal_dc_17['month'],
         normal_dc_17['CO'],linewidth='3')
plt.plot(delay_dc_17['month'],
         delay_dc_17['CO'],'.g',markersize=15)
plt.plot(cancel_dc_17['month'],
         cancel_dc_17['CO'],'.m',markersize=15)
plt.title('2017 temp')
plt.legend(['normal_temp','delay_temp','cancel_temp'],prop={'size':15})

plt.subplot(313)
plt.plot(normal_dc_18['month'],
         normal_dc_18['CO'],linewidth='3')
plt.plot(delay_dc_18['month'],
         delay_dc_18['CO'],'.g',markersize=15)
plt.plot(cancel_dc_18['month'],
         cancel_dc_18['CO'],'.m',markersize=15)
plt.title('2018 temp')
plt.legend(['normal_temp','delay_temp','cancel_temp'],prop={'size':15}) # bbox_to_anchor=(1, 0.5)


# plt.savefig('C:\\python_DA\\semi\\result\\indc1618_11elements_3label\\1618_temp.png')
```
